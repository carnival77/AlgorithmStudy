```python
# 미세먼지 안녕
R,C,T = map(int,input().split())
board = [list(map(int,input().split())) for _ in range(R)]
after_t = 0
dx = [1,-1,0,0]
dy = [0,0,1,-1]

air_machine = []
# 공기 청정기 위치
for x in range(R):
    for y in range(C):
        if board[x][y] == -1:
            air_machine.append([x,y])
air_machine.sort()
while True:
    # 1초 경과
    after_t += 1
    # 미세먼지 위치
    dirt = []
    for x in range(R):
        for y in range(C):
            if board[x][y] >= 1:
                dirt.append([x,y])
    # 미세먼지 확산
    plus_dirt = []
    minus_dirt = []
    for x,y in dirt:
        for i in range(4):
            nx = x+dx[i]
            ny = y+dy[i]
            new_dirt=board[x][y]//5
            if nx>=0 and ny>=0 and nx<R and ny<C and board[nx][ny] != -1:
                plus_dirt.append([new_dirt,nx,ny])
                minus_dirt.append([new_dirt,x,y])
    for new_dirt,x,y in plus_dirt:
        board[x][y] += new_dirt
    for new_dirt,x,y in minus_dirt:
        board[x][y] -= new_dirt
    # 위쪽 바람 불기
    x,y=air_machine[0]
    dirt_list = []
    if x >= 1:
        board[x-1][y] = 0
    for ny in range(1,C):
        if board[x][ny] > 0:
            if ny == C-1:
                dirt_list.append([x-1,ny,board[x][ny]])
            else:
                dirt_list.append([x,ny+1,board[x][ny]])
            board[x][ny] = 0
    for nx in range(x,-1,-1):
        if board[nx][C-1] > 0:
            if nx == 0:
                dirt_list.append([nx,C-2,board[nx][C-1]])
            else:
                dirt_list.append([nx-1,C-1,board[nx][C-1]])
            board[nx][C-1] = 0
    for ny in range(C-1,-1,-1):
        if board[0][ny] > 0:
            if ny == 0:
                dirt_list.append([1,ny,board[0][ny]])
            else:
                dirt_list.append([0,ny-1,board[0][ny]])
            board[0][ny] = 0
    for nx in range(x-1):
        if board[nx][0] > 0:
            dirt_list.append([nx+1,0,board[nx][0]])
        board[nx][0] = 0
        
    # 아래 바람 불기
    x,y=air_machine[1]
    if x<=R-2:
        board[x+1][y] = 0
    for ny in range(1,C):
        if board[x][ny] > 0:
            if ny == C-1:
                dirt_list.append([x+1,ny,board[x][ny]])
            else:
                dirt_list.append([x,ny+1,board[x][ny]])
            board[x][ny] = 0
    for nx in range(x,R):
        if board[nx][C-1] > 0:
            if nx == R-1:
                dirt_list.append([nx,C-2,board[nx][C-1]])
            else:
                dirt_list.append([nx+1,C-1,board[nx][C-1]])
            board[nx][C-1] = 0
    for ny in range(C-1,-1,-1):
        if board[R-1][ny] > 0:
            if ny == 0:
                dirt_list.append([R-2,ny,board[R-1][ny]])
            else:
                dirt_list.append([R-1,ny-1,board[R-1][ny]])
            board[R-1][ny] = 0
    for nx in range(R-1,x+1,-1):
        if board[nx][0] > 0:
            dirt_list.append([nx-1,0,board[nx][0]])
            board[nx][0] = 0
    for x,y,new_dirt in dirt_list:
        board[x][y] = new_dirt 

    if after_t == T:
        break
answer = 0
for i in range(R):
    for j in range(C):
        if board[i][j] >=0:
            answer+=board[i][j]
print(answer)
```

```python
# 원판 돌리기

N,M,T = map(int,input().split())
board = [list(map(int,input().split())) for _ in range(N)]
rotations = [list(map(int,input().split())) for _ in range(T)]
dx = [1,-1,0,0]
dy = [0,0,1,-1]
def rot_left(a_list):
    a=a_list.pop()
    a_list.insert(0,a)
    return a_list
def rot_right(a_list):
    a=a_list.pop(0)
    a_list.append(a)
    return a_list

while rotations:
    x,d,k = rotations.pop(0)
    ro_won = []
    # 회전할 원판 고르기
    for i in range(1,N+1):
        if x*i <= N:
            ro_won.append(x*i-1)
        else:
            break
    # 원판 회전시키기
    for won in ro_won:
        if d == 0:
            for _ in range(k):
                board[won] = rot_left(board[won])
        else:
            for _ in range(k):
                board[won] = rot_right(board[won])
    # 인접한 수 찾기
    adj = []
    for x in range(N):
        for y in range(M):
            if board[x][y] !=0:
                for i in range(4):
                    nx = x+dx[i]
                    ny = (y+dy[i])%M
                    if nx < 0 or nx>=N:
                        continue
                    if board[x][y] == board[nx][ny]:
                        adj.append([x,y])
                        adj.append([nx,ny])
    if adj:
        # 인접한 수 0으로 만들기
        for x,y in adj:
            board[x][y] = 0
    else:
        # 적힌 수의 평균 구하기
        total = 0
        total_len = 0
        for x in range(N):
            for y in range(M):
                if board[x][y] != 0:
                    total+=board[x][y]
                    total_len+=1
        # 평균보다 작은거 빼기
        for x in range(N):
            for y in range(M):
                if board[x][y] != 0:
                    if board[x][y]> (total/total_len):
                        board[x][y] -= 1
                    elif board[x][y]< (total/total_len):
                        board[x][y] += 1
answer = 0
for x in range(N):
    for y in range(M):
        answer += board[x][y]
print(answer)
```

```python
# 컨베이어 벨트 위의 로봇

n,k = map(int,input().split())
board = list(map(int,input().split()))

def rotation(board):
    a = board.pop()
    board.insert(0,a)
    return board

robot = [0]*n

answer = 0

while True:
    board=rotation(board)
    robot=rotation(robot)
    robot[n-1] = 0
    for i in range(n-2,-1,-1):
        if robot[i] == 1 and robot[i+1] == 0 and board[i+1] != 0:
            board[i+1] -= 1
            robot[i],robot[i+1] = robot[i+1],robot[i]
        robot[n-1] = 0
    if board[0] != 0 and robot[0] != 1:
        robot[0] = 1
        board[0] -= 1
    answer +=1
    ans = 0
    for i in board:
        if i == 0:
            ans += 1
    if ans >= k:
        break
print(answer)
```

```python
# 마법사 상어와 비바라기

from collections import deque
N,M = map(int,input().split())
board = [list(map(int,input().split())) for _ in range(N)]
go = deque()
for _ in range(M):
    go.append(list(map(int,input().split())))
cloud = [[N-1,0],[N-1,1],[N-2,0],[N-2,1]]
dx = [0,-1,-1,-1,0,1,1,1]
dy = [-1,-1,0,1,1,1,0,-1]
dx_1 = [1,-1,1,-1]
dy_1 = [1,1,-1,-1]
while go:
    d,s = go.popleft()
    d -= 1
    new_cloud = []
    visited = [[False]*N for _ in range(N)]
    for x,y in cloud:
        nx = (x+dx[d]*s) % N
        ny = (y+dy[d]*s) % N
        new_cloud.append([nx,ny])
    for x,y in new_cloud:
        board[x][y] +=1
        visited[x][y] = True
    # 물 복사 버그
    for x,y in new_cloud:
        an = 0
        for i in range(4):
            nx = x+dx_1[i]
            ny = y+dy_1[i]
            if nx<0 or nx>=N or ny <0 or ny>=N:
                continue
            if board[nx][ny] != 0:
                an+=1
        board[x][y] += an
    cloud = []
    # 구름 생성
    for x in range(N):
        for y in range(N):
            if board[x][y] >=2:
                if not visited[x][y]:
                    board[x][y] -= 2
                    cloud.append([x,y])
answer =0
for x in range(N):
    for y in range(N):
        answer += board[x][y]
print(answer)
```

```python
# 마법사 상어와 파이어볼
N,M,K = map(int,input().split())
board = [[[] for _ in range(N)] for _ in range(N)]
fire = []
for _ in range(M):
    r,c,m,s,d = map(int,input().split())
    fire.append([r-1,c-1,m,s,d])
dx = [-1,-1,0,1,1,1,0,-1]
dy = [0,1,1,1,0,-1,-1,-1]

end = 0
while end < K:
    end += 1
    while fire:
        x,y,m,s,d = fire.pop(0)
        nx = (x+dx[d]*s) % N
        ny = (y+dy[d]*s) % N
        board[nx][ny].append([m,s,d])
    # 파이어볼 합친 후 나누기
    for x in range(N):
        for y in range(N):
            if len(board[x][y]) > 1:
                weight = 0
                speed = 0
                d_list = []
                fire_len = len(board[x][y])
                while board[x][y]:
                    m,s,d = board[x][y].pop()
                    weight+=m
                    speed+=s
                    d_list.append(d)
                weight = weight//5
                if weight != 0:
                    cnt_odd = 0
                    cnt_even= 0
                    for d in d_list:
                        if d%2 == 1:
                            cnt_odd+=1
                        if d%2 == 0:
                            cnt_even += 1
                    if cnt_odd == fire_len or cnt_even == fire_len:
                        for i in range(4):
                            fire.append([x,y,int(weight),int(speed//fire_len),i*2])
                    else:
                        for i in range(4):
                            fire.append([x,y,int(weight),int(speed//fire_len),i*2+1])
            elif len(board[x][y]) == 1:
                m,s,d=board[x][y].pop(0)
                fire.append([x,y,m,s,d])

answer = 0                 
for x,y,m,s,d in fire:
    answer += m
print(answer)
```
